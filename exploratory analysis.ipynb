```python
from matplotlib.patches import FancyArrowPatch
import matplotlib.patches as mpatches

top5_artists = spotify_data.groupby('artist_names')['streams'].sum().sort_values(ascending=False).head(5).index
features = ['danceability', 'energy', 'speechiness', 'acousticness', 'valence', 'tempo']

# Normalize tempo to 0-1
df = spotify_data.copy()
df['tempo'] = (df['tempo'] - df['tempo'].min()) / (df['tempo'].max() - df['tempo'].min())

angles = np.linspace(0, 2 * np.pi, len(features), endpoint=False).tolist()
angles += angles[:1]

fig, axes = plt.subplots(1, 5, figsize=(20, 4), subplot_kw=dict(polar=True))
colors = ['#e1306c', '#833ab4', '#f77737', '#ff0087', '#405de6']

for ax, artist, color in zip(axes, top5_artists, colors):
    values = df[df['artist_names'] == artist][features].mean().tolist()
    values += values[:1]
    ax.plot(angles, values, color=color, linewidth=2)
    ax.fill(angles, values, color=color, alpha=0.25)
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(features, size=7)
    ax.set_title(artist[:15], size=9, fontweight='bold', pad=10)
    ax.set_ylim(0, 1)

plt.suptitle('Audio Profile of Top 5 Artists', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Select features and target
features = ['danceability', 'energy', 'loudness', 'valence', 'tempo', 'days_on_chart']
target = 'streams'

X = spotify_data[features].values
y = spotify_data[target].values

# Normalize X (important!)
X = (X - X.mean(axis=0)) / X.std(axis=0)

# Add bias column (column of 1s)
X = np.hstack([np.ones((X.shape[0], 1)), X])

# Train/test split manually
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# ── Gradient Descent ──────────────────────────────
learning_rate = 0.01
epochs = 500
m = len(X_train)
theta = np.zeros(X_train.shape[1])
losses = []

for epoch in range(epochs):
    predictions = X_train @ theta
    error = predictions - y_train
    gradient = (1/m) * X_train.T @ error
    theta -= learning_rate * gradient
    loss = (1/(2*m)) * np.sum(error**2)
    losses.append(loss)

# ── Evaluation ────────────────────────────────────
y_pred = X_test @ theta
ss_res = np.sum((y_test - y_pred) ** 2)
ss_tot = np.sum((y_test - y_test.mean()) ** 2)
r2 = 1 - ss_res / ss_tot
rmse = np.sqrt(np.mean((y_test - y_pred) ** 2))

print(f"R² Score: {r2:.4f}")
print(f"RMSE:     {rmse:,.0f} streams")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Loss curve
axes[0].plot(losses, color='#e1306c', linewidth=2)
axes[0].fill_between(range(len(losses)), losses, alpha=0.2, color='#e1306c')
axes[0].set_title('Loss over Training Epochs', fontweight='bold')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('MSE Loss')

# Predicted vs Actual (sample 500 points so it's readable)
sample = np.random.choice(len(y_test), 500, replace=False)
axes[1].scatter(y_test[sample], y_pred[sample], alpha=0.4, color='#833ab4', s=15)
axes[1].plot([y_test.min(), y_test.max()], 
             [y_test.min(), y_test.max()], 
             color='#e1306c', linewidth=2, linestyle='--', label='Perfect prediction')
axes[1].set_title('Predicted vs Actual Streams', fontweight='bold')
axes[1].set_xlabel('Actual Streams')
axes[1].set_ylabel('Predicted Streams')
axes[1].legend()

plt.suptitle(f'Linear Regression from Scratch — R² = {r2:.4f}', 
             fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```
