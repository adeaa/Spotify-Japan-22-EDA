```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-v0_8-darkgrid')
colors=plt.cm.RdPu(np.linspace(0.2, 0.8, 10))
```

# DataFrame info 

```python
spotify_data = pd.read_csv("top200jp2022_daily.csv")
spotify_data.head()
```

```python
spotify_data.info()
```

```python
spotify_data.describe(
)
```

## Top artists

```python
top_artists = spotify_data.groupby('artist_names')['streams'].sum().sort_values(ascending=False).head(10)
top_artists
```

```python
top_artists.index
```

```python
fig , ax = plt.subplots(figsize=(12,6))
ax.barh(width= top_artists.values,y=top_artists.index,color=colors[::-1])
ax.set(title="Top 10 Most Streamed Artists in Japan (2022)",xlabel="Total Streams", ylabel="Artist Name");
```

## What makes a song stay on the chart longer?

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for ax, feature in zip(axes, ['danceability', 'energy', 'valence']): #valence refers to the emotional tone or "positivity" conveyed by a track
    ax.scatter(x=spotify_data[feature], y=spotify_data['days_on_chart'], alpha=0.1, color=colors[8],s=1)
    ax.set(xlabel=feature,ylabel='Days on Chart',title=(f'{feature} vs Longevity'))

plt.suptitle('Audio Features vs Chart Longevity', fontsize=14, fontweight='bold');
```

## Genre distribution

```python
spotify_data.head()
```

```python
top_genres = spotify_data['artist_genre'].value_counts().head(10)

plt.figure(figsize=(10, 6))
plt.pie(top_genres.values, labels=top_genres.index, autopct='%1.1f%%',
        colors=colors)
plt.title('Top Genres in Japan Spotify Charts 2022', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
spotify_data.info()
```

```python
features = ['danceability', 'energy', 'loudness','key', 'speechiness', 
            'acousticness', 'valence', 'tempo', 'streams','instrumentalness']
```

```python
corr = spotify_data[features].corr()
corr
```

```python
fig, ax = plt.subplots(figsize=(12, 9))
im = ax.imshow(corr, cmap='RdPu')
plt.colorbar(im)
ax.set_xticks(range(len(features)))
ax.set_yticks(range(len(features)))
ax.set_xticklabels(features, rotation=45, ha='right')
ax.set_yticklabels(features)

for i in range(len(features)):
    for j in range(len(features)):
        ax.text(j, i, f'{corr.iloc[i, j]:.2f}', ha='center', va='center', fontsize=8)

plt.title('Audio Feature Correlations', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
spotify_data['date'] = pd.to_datetime(spotify_data['date'])
monthly = spotify_data.groupby(spotify_data['date'].dt.month)['streams'].mean()

plt.figure(figsize=(12, 5))
plt.plot(monthly.index, monthly.values, color='#e1306c', linewidth=2.5, marker='o')
plt.fill_between(monthly.index, monthly.values, alpha=0.2, color='#e1306c')
plt.xlabel('Month')
plt.ylabel('Average Streams')
plt.title('Average Streams by Month (2022)', fontsize=14, fontweight='bold')
plt.xticks(range(1, 13), ['Jan','Feb','Mar','Apr','May','Jun',
                            'Jul','Aug','Sep','Oct','Nov','Dec']);
```

```python
import japanize_matplotlib

longest = spotify_data.groupby('track_name')['days_on_chart'].max().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
plt.barh(longest.index, longest.values, color=colors)
plt.xlabel('Days on Chart')
plt.title('Top 10 Longest Charting Songs in Japan (2022)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
